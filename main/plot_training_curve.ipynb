```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import pyplot
import torch
```

```python
def train_plot(losses):
    epochs = range(0, len(losses))
    plt.plot(epochs, losses, label='Training Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.savefig('Training Loss.png')
```

```python
def test_plot(errors):
    epochs = range(0, len(errors))
    plt.plot(epochs, errors, label='Validation Error')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.savefig('Validation Error.png')
```
