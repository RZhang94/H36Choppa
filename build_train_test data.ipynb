```python
import numpy as np
import os
```

```python
# check the subject in traning data
dataset_path = '/content/drive/MyDrive/HME/Temporal-Transformer/trainingdata.npy'
# args.dataset should be the name of the npz file

data = np.load(dataset_path, allow_pickle=True)
subject_name = ''

for sample in data:
    # print(sample[0][0])

    if subject_name == sample[0][0]:
        continue
    else:
        subject_name = sample[0][0]
        print(subject_name)
# So according to the names printed, the training set only contains S1 S5 S7 S11
# So the test subjects should be S2 S6 S8
```

```python
# the test set
data = []
subjects = ['S6','S8']
root_folder = '/content/drive/MyDrive/HME/H36M Data/Cropped Images/HMR Outputs/Structure 2_gt'
for subject in subjects:
    folder_name = os.path.join(root_folder, subject)
    # print(folder_name)
    for filename in os.listdir(folder_name):
        # import pdb
        # pdb.set_trace()
        # print(filename)
        file_path = os.path.join(folder_name, filename)
        # print(file_path)
        data.append(np.load(file_path, allow_pickle=True).transpose(1,0))
        # print(len(data))

# print(len(data))
print(data[0])
```

```python
test_data = data[0]
for i in data[1:]:
    test_data = np.vstack([test_data, i])

for sample in test_data:
    print(sample[0])
```
